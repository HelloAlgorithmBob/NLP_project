```python
from transformers import pipeline

# Load the pipeline for question answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad", tokenizer="distilbert-base-cased")

# Reading passage
passage = """
The Industrial Revolution was a period of major industrialization that took place during the late 1700s and early 1800s.
It marked a significant turning point in history as it introduced the use of machinery, factories, and new manufacturing processes.
This era had a profound impact on various aspects of society, including the economy, agriculture, and living conditions.
"""

# List of questions
questions = [
    "When did the Industrial Revolution occur?",
    "What were some of the key changes during this period?",
    "How did the Industrial Revolution impact society?",
]

# Answer the questions using the model
for question in questions:
    result = qa_pipeline(question=question, context=passage)
    print(f"Question: {question}")
    print(f"Answer: {result['answer']}")
    print(f"Confidence: {result['score']}")
    print("=" * 50)
```
